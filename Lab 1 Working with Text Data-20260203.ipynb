```python
# Extract relevant information from an XML document. 
import xml.etree.ElementTree as ET 
 
# Parse the XML file 
try:
    tree = ET.parse('news.xml') 
    root = tree.getroot() 
     
    # It's better to open the file once and do all operations inside
    with open('stored_titles.txt', 'w', encoding='utf-8') as file: 
        for article in root.findall('article'): 
            title_element = article.find('title')
            
            # Check if title exists to avoid errors on empty tags
            if title_element is not None:
                title = title_element.text
                print("Extracted Title:", title)
                file.write(title + '\n')
                
    print("\nExtraction complete. Titles saved to 'stored_titles.txt'.")

except FileNotFoundError:
    print("Error: 'news.xml' not found. Please ensure the file exists in the directory.")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
# First, ensure you have the library installed
# !pip install PyPDF2

import PyPDF2

def extract_pdf_to_txt(pdf_path, output_path):
    try:
        # Open the PDF file in read-binary mode
        with open(pdf_path, 'rb') as pdf_file:
            # Create a PDF reader object
            reader = PyPDF2.PdfReader(pdf_file)
            
            # Initialize a list to hold text from all pages
            all_text = []
            
            # Loop through each page in the PDF
            for page_num in range(len(reader.pages)):
                page = reader.pages[page_num]
                text = page.extract_text()
                
                # Append the text to our list
                if text:
                    all_text.append(f"--- Page {page_num + 1} ---\n{text}\n")
            
            # Join all extracted text into one string
            final_content = "\n".join(all_text)
            
            # Save the content to the .txt file
            with open(output_path, 'w', encoding='utf-8') as txt_file:
                txt_file.write(final_content)
                
        print(f"Success! Text extracted to {output_path}")

    except FileNotFoundError:
        print("The PDF file was not found. Please check the file path.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Run the function
extract_pdf_to_txt('Business_Proposal.pdf', 'business_proposal_all.txt')
```

```python
# !pip install PyPDF2  # Run this if you haven't installed the library yet

import PyPDF2

def extract_specific_page(pdf_path, output_path, page_number):
    try:
        # Open the PDF file
        with open(pdf_path, 'rb') as pdf_file:
            reader = PyPDF2.PdfReader(pdf_file)
            
            # Check if the page exists (total pages vs requested page)
            if len(reader.pages) >= page_number:
                # Page index is page_number - 1 (e.g., Page 2 is index 1)
                page = reader.pages[page_number - 1]
                text = page.extract_text()
                
                if text.strip():
                    # Save to the text file
                    with open(output_path, 'w', encoding='utf-8') as txt_file:
                        txt_file.write(text)
                    print(f"Success! Page {page_number} extracted and saved to {output_path}")
                else:
                    print(f"Page {page_number} was found but appears to be empty or an image.")
            else:
                print(f"Error: The PDF only has {len(reader.pages)} pages.")

    except FileNotFoundError:
        print("Error: 'Business_Proposal.pdf' not found. Ensure the file is in the same folder as this notebook.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

# Run the function for Page 2
extract_specific_page('Business_Proposal.pdf', 'business_proposal_page_2.txt', 2)
```
